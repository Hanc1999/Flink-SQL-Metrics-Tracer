## Euclidean Distance and Cosine Similarities Pair Comparing

```python
def calc_pair_distances(embeddings):
    l = len(embeddings)
    res = []
    for i in range(l):
        v1 = embeddings[i]
        tmp = []
        for j in range(l):
            v2 = embeddings[j]
            dis = np.linalg.norm(v1 - v2)
            tmp.append(dis)
        res.append(tmp)
    res = np.array(res)
    return res

def calc_pair_cossim(embeddings):
    embeddings = normalize(embeddings)
#     cos_sim = dot(a, b)/(norm(a)*norm(b))
    l = len(embeddings)
    res = []
    for i in range(l):
        v1 = embeddings[i]
        tmp = []
        for j in range(l):
            v2 = embeddings[j]
            sim = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
            tmp.append(sim)
        res.append(tmp)
    res = np.array(res)
    return res

def reorder_as(order, keys, embeddings): # reorder rows
    target = []
    for o in order:
        target.append(keys.index(o))
#     print(target)
    res = embeddings[target, :]
    return res

# for demo
def fig_demo(m):
    m *= 255.0/m.max()
    plt.matshow(m)
    plt.colorbar()
    plt.show()

def compare_2_demo(m1, m2):
    the_max = max(m1.max(), m2.max())
    
    plt.imshow(m1*255.0/m1.max(), cmap='gray')
    plt.colorbar()
    plt.show()

    plt.imshow(m2*255.0/m2.max(), cmap='gray')
    plt.colorbar()
    plt.show()
    
def compare_mul_demo(ms):
    miner = min([m.min() for m in ms])
    maxer = max([m.max() for m in ms])
    
    l = len(ms)
    w = 4
    h = l // 4 + (1 - (l%4 == 0))
    
    fig, axs = plt.subplots(h, min(w, l), figsize=(10, 10))
    
    for m1_i in range(l):
        m1 = ms[m1_i]
        if h == 1:
            ax = axs[m1_i]
        else:
            ax = axs[m1_i // w, m1_i % w]
#         ax.imshow((m1-m1.min())/(m1.max()-m1.min())*255.0, cmap='gray')
        ax.imshow((m1-miner)/(maxer-miner)*255.0, cmap='gray')
#     plt.colorbar()
    plt.show()
```
